```python
minnesota = read_csv('/content/drive/MyDrive/STAT 4680/STAT 4680 - Team Project!/minnesota_stats_newvar.csv')
player_ids = minnesota['PLAYER_ID'].unique()
min_id = minnesota['TEAM_ID'].unique()
min_id = min_id[0]
seasons = minnesota['SEASON_ID'].unique()
```

```python
minnesota_analysis = minnesota[['PLAYER_ID', 'SEASON_ID', 'TEAM_ID']]
```

```python
leaguedashplayerstats.LeagueDashPlayerStats(season_type_all_star = 'Regular Season',
                                            measure_type_detailed_defense = ['Defense'],
                                            season = seasons[0],
                                            team_id_nullable=min_id,
                                            per_mode_detailed = ["Per100Possessions"]).get_data_frames()[0]
```

```python
defense_stats = leaguedashplayerstats.LeagueDashPlayerStats(season_type_all_star = 'Regular Season',
                                                                       measure_type_detailed_defense = ['Defense'],
                                                                       season = seasons[0],
                                                                       team_id_nullable=min_id,
                                                                       per_mode_detailed = ["Per100Possessions"]).get_data_frames()[0]

defense_stats['SEASON_ID'] = seasons[0]

for s in seasons[1:]:
  tempsingle = leaguedashplayerstats.LeagueDashPlayerStats(season_type_all_star = 'Regular Season',
                                                           measure_type_detailed_defense = ['Defense'],
                                                           season = s,
                                                           team_id_nullable=min_id,
                                                           per_mode_detailed = ["Per100Possessions"]).get_data_frames()[0]
  tempsingle['SEASON_ID'] = s

  defense_stats = pd.concat([tempsingle, defense_stats])
```

```python
defense_stats.to_csv('/content/drive/MyDrive/STAT 4680/STAT 4680 - Team Project!/defense-stats.csv')
files.download('/content/drive/MyDrive/STAT 4680/STAT 4680 - Team Project!/defense-stats.csv')
```
