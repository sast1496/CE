```python
shot_chart_avg = shot_chart.groupby(['SEASON_ID', 'PLAYER_ID'], as_index=False)['SHOT_DISTANCE'].mean()
player_stats = player_stats.loc[player_stats['PLAYER_ID'].isin(shot_chart_avg['PLAYER_ID'])]

minnesota_stats = player_stats.merge(shot_chart_avg,
                   left_on = ["PLAYER_ID", "SEASON_ID"],
                   right_on = ["PLAYER_ID", "SEASON_ID"],
                   how = "left",
                   suffixes=["",""])
```

```python
player_bio_stats = player_bio_stats[['PLAYER_ID','SEASON_ID','PLAYER_HEIGHT_INCHES', 'PLAYER_WEIGHT', 'NET_RATING']]

minnesota_stats = minnesota_stats.merge(player_bio_stats,
                   left_on = ["PLAYER_ID", "SEASON_ID"],
                   right_on = ["PLAYER_ID", "SEASON_ID"],
                   how = "left",
                   suffixes=["",""])
```

```python
hustle_stats = hustle_stats[['PLAYER_ID', 'SEASON_ID', 'CONTESTED_SHOTS']]

minnesota_stats = minnesota_stats.merge(hustle_stats,
                   left_on = ["PLAYER_ID", "SEASON_ID"],
                   right_on = ["PLAYER_ID", "SEASON_ID"],
                   how = "left",
                   suffixes=["",""])

minnesota_stats["AVG_CONTESTED_SHOTS"] = minnesota_stats["CONTESTED_SHOTS"]/minnesota_stats['GP']
minnesota_stats["CONTESTED_SHOTS"]=minnesota_stats['AVG_CONTESTED_SHOTS']
minnesota_stats = minnesota_stats.drop(columns = ['AVG_CONTESTED_SHOTS'])

```

```python
minnesota_stats = minnesota_stats.rename(columns={"PLAYER_HEIGHT_INCHES": "HEIGHT", "PLAYER_WEIGHT": "WEIGHT"})
```

```python
advanced_stats_filtered = advanced_stats[['PLAYER_ID','SEASON_ID','OFF_RATING', 'DEF_RATING', 'PIE', 'USG_PCT', 'POSS']]

minnesota_stats = minnesota_stats.merge(advanced_stats_filtered,
                   left_on = ["PLAYER_ID", "SEASON_ID"],
                   right_on = ["PLAYER_ID", "SEASON_ID"],
                   how = "left",
                   suffixes=["",""])
```

```python
defensive_stats_filtered = defensive_stats[['PLAYER_ID','SEASON_ID','DEF_WS']]

minnesota_stats = minnesota_stats.merge(defensive_stats_filtered,
                   left_on = ["PLAYER_ID", "SEASON_ID"],
                   right_on = ["PLAYER_ID", "SEASON_ID"],
                   how = "left",
                   suffixes=["",""])
```

```python
team_roster_filtered = team_roster[['PLAYER_ID', 'SEASON_ID', 'POSITION']]

minnesota_stats = minnesota_stats.merge(team_roster_filtered,
                   left_on = ["PLAYER_ID", "SEASON_ID"],
                   right_on = ["PLAYER_ID", "SEASON_ID"],
                   how = "left",
                   suffixes=["",""])
```

```python
minnesota_stats['EFF'] = (minnesota_stats['PTS']
                          + minnesota_stats['REB']
                          + minnesota_stats['AST']
                          + minnesota_stats['STL']
                          - minnesota_stats['BLK']
                          - minnesota_stats['FGM']
                          - minnesota_stats['FTM']
                          - minnesota_stats['TOV']) / minnesota_stats['GP']
```
